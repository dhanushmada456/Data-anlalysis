```python
import numpy as np
import matplotlib.pyplot as plt
runs=np.array([100,50,91,78,89,25,34,19,9,10])
w=np.array([1,0,2,0,3,7,8,9,7,5])
plt.scatter(w,runs,color='black',label='India')#scatter graph
plt.title('IndvsAus_score')
plt.legend(loc='best')#it will locate best position for labelsal = np.array([12000,60000,80000,76000])
plt.show()#instead of print we use .show in matplotlib
```

```python
tigar=np.linspace(-2*np.pi, 2*np.pi, 100)
print(tigar)
plt.plot(tigar, np.sin(tigar) ,color='black',linestyle='dotted')
plt.title("sin(x)")
plt.show()
```

```python
sal = np.array([12000,60000,80000,76000])
wd = np.array([210,300,230,256])
years = [1,2]
bonus=[]
for i,j in zip(sal,wd):
  if j > 250:
    i = i+i*0.1
    bonus.append(i)
  else:
    i = i-i*0.1
    bonus.append(i)
print(bonus)
```

```python
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)#
plt.plot(years,profit_a,color='hotpink',linewidth = '8',label =
'CompanyA',marker='.',ms='20',mec='k')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label =
'CompanyB',marker = 'H')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.array([25,60,5,10])
labe = ["AIML","Python","Pandas","Numpy"]
color = ['green','pink','coral','yellow']
plt.pie(a,labels = labe,colors= color)
plt.show()
```
