```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.theguardian.com/news/datablog/2012/may/04/world-top-100-airports'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
title = soup.find('title')
print(title)
print(title.text)
```

```python
table = soup.find('table')
print(table)
```

```python
Headings = table.find_all('th')
print(Headings)
```

```python
Header = [i.text for i in Headings]
print(Header)
```

```python
import pandas as pd
df = pd.DataFrame(columns = Header)
df
```

```python
rows = table.find_all('tr')
print(rows)
print(df.columns)
```

```python
ind = 1
for i in rows[1:100]:
    row_data = i.find_all('td')
    every_row_data = [i.text for i in row_data]

    # Check for mismatched columns
    if len(df.columns) != len(every_row_data):
        # Handle the mismatch here (e.g., adjust columns or data types)
        pass

    # Assign values to DataFrame row using column names
    df.loc[ind] = dict(zip(df.columns, every_row_data))

    ind += 1
    print(every_row_data)
```
